```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# Continous variable in 4 different ways

x = tips['total_bill']
sns.distplot(x)
plt.title('Distribution of total bill')
```

```python
x = tips['total_bill']
sns.boxplot(x)
```

```python
x = tips['total_bill']
sns.kdeplot(x)
```

```python
x = tips['total_bill']
sns.distplot(x, kde=False, rug=True);
```

```python
# Two continous variables in 3 different ways

g = sns.lmplot(y='tip', 
               x='total_bill',
               data=tips, 
               fit_reg=True, 
               scatter_kws={'alpha':0.4},
               palette='pastel')
g.set_ylabels("Tips")
g.set_xlabels("Total Bill")
plt.title('Meal Tips by Total Bill')
plt.show()
```

```python
g = sns.lmplot(y='tip', 
               x='total_bill',
               data=tips, 
               fit_reg=False, 
               scatter_kws={'alpha':0.4},
               palette='pastel')
g.set_ylabels("Tips")
g.set_xlabels("Total Bill")
plt.title('Meal Tips by Total Bill')
plt.show()
```

```python

x = tips['total_bill']
y = tips['tip']

bins = np.linspace(-10, 10, 100)

plt.hist(x, bins, alpha=0.5, label='Tips')
plt.hist(y, bins, alpha=0.5, label='Total Bill')
plt.legend(loc='upper right')
plt.title('Distribution of Tips and Total Bill')
plt.show()

```

```python
# One continous variable and two categorical variables in 4 different ways


tips = sns.load_dataset("tips")
tips['percentage'] = tips['tip'] / tips['total_bill']
tips['odd?'] = tips['size'] % 2

def create_factor_plot(x_axis, y_axis, x_label, y_label, dataframe):
    sns.set(style="whitegrid")
    g = sns.factorplot(x=x_axis, y=y_axis, data=dataframe,
                       size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
    g.despine(left=True)
    g.set_ylabels(y_label)
    g.set_xlabels("")
    plt.title("Pointplot: " + y_label + " by " + x_label)
    plt.show()
    
def create_lmplot_tip_vs_total_bill_by(criterion, criterion_name):
    g = sns.lmplot(y='tip', # Variable 1.
               x='total_bill', # Variable 2.
               hue=criterion,
               data=tips, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
    g.set_ylabels("Tip")
    g.set_xlabels("Total Bill")
    plt.title("Scatterplot: Meal Tip Percentages by Total Bill and " + criterion_name)
    plt.show()

ax = sns.boxplot(x='time',y='percentage',hue='sex',data=tips) 
plt.title('Plot 2: Meal Tip Percentages by Gender and Time')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tip Percentage')
plt.show()

sns.set(style="darkgrid")
g = sns.factorplot(x="sex", y="percentage", hue="time", data=tips,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Tip Percentages")
g.set_xlabels("")
plt.title('Barplot: Meal Tip Percentages by Time and Gender')
plt.show()

sns.set(style="whitegrid")
g = sns.factorplot(x="sex", y="percentage", hue="time", data=tips,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Tip Percentages")
g.set_xlabels("")
plt.title('Pointplot: Meal Tip Percentages by Time and Gender')
plt.show()

g = sns.FacetGrid(tips, row="sex", col="time")
g.map(plt.hist, "percentage", color="steelblue",  lw=0)
plt.subplots_adjust(top=0.9)
g.set_axis_labels('Tip in Percentage')
plt.suptitle('Plot 1: Meal Tip Percentages by Gender and Time')
sns.despine(trim=True)
plt.show()

```
